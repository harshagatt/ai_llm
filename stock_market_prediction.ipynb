```python
# remove the last 2 columns
del sp500["Dividends"]
del sp500["Stock Splits"]
# setup the target
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
#view the dataframe
sp500
```

```python
#create a new column called target and check when next day opening was greater than last day closing

sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
#remove the data before 1990
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
#training initial model using RandomForestClassifier
#realtive quick, pickup non linear data relationship
from sklearn.ensemble import RandomForestClassifier

#n - number of individual decision tree
#min_sample_split - protect against overfitting, higher the number, lesser the overfitting
#random_state - random numbers generated are in predicitable sequence
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

#train and test test for time series data
#leaking information into the model using future data
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

#state the predictors
predictors = ["Close", "Volume", "Open", "High", "Low"]

#fit the model using predictors columns and train the Target column
model.fit(train[predictors], train["Target"])


```

```python
#measure the accuracy of the model
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])

import pandas as pd
#turn into a pandas series
preds = pd.Series(preds, index=test.index)
precision_score(test["Target"], preds)
#
```

```python
# prediction was 53% correct
#concate actual values and predicted values
#
combined = pd.concat([test["Target"], preds], axis=1)
# plot this data
combined.plot()
```

```python
#build a backtesting system

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds,
                      index=test.index,
                      name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined

```

```python
# create a backtest function
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)

```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#benchmark the data
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# adding additional predictors to the model to check if it improves the accuracy

horizons = [2,5,60,250,1000,1250]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors+= [ratio_column, trend_column]

```

```python
#drop the extra columns
sp500 = sp500.dropna()
```

```python
#update the model and increase the estimators to 200
#reduce the min sample split
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)


```

```python
#rewrite the predictor function
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.65] = 1
    preds[preds <.65] = 0
    preds = pd.Series(preds,
                      index=test.index,
                      name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#update the predictions and do the backtest
predictions = backtest(sp500, model, new_predictors)
```

```python
#do the value count again
predictions["Predictions"].value_counts()
```

```python
#check the accuracy score
precision_score(predictions["Target"], predictions["Predictions"])
```
